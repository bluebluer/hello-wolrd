```python
import pandas as pd
```

```python
euro12 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv', sep=',')
euro12
```

```python
# Select only the Goal column.
euro12.Goals
```

```python
# How many team participated in the Euro2012?
euro12.shape[0]
```

```python
# What is the number of columns in the dataset?
euro12.info()
```

```python
# View only the columns Team, Yellow Cards and Red Cards
# and assign them to a dataframe called discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
discipline
```

```python
# Sort the teams by Red Cards, then to Yellow Cards
discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending = False)
```

```python
# Calculate the mean Yellow Cards given per Team
round(discipline['Yellow Cards'].mean())
```

```python
# Filter teams that scored more than 6 goals
euro12[euro12.Goals > 6]
```

```python
# Select the teams that start with G
euro12[euro12.Team.str.startswith('G')]
```

```python
# Select the first 7 columns
# use .iloc to slices via the position of the passed integers
# : means all, 0:7 means from 0 to 7

euro12.iloc[: , 0:7]
```

```python
# Select all columns except the last 3.
# use negative to exclude the last 3 columns

euro12.iloc[: , :-3]
```

```python
# Present only the Shooting Accuracy from England, Italy and Russia
# loc是location的意思，和iloc中i的意思是指integer，所以它只接受整数作为参数，详情见下面。
euro12.loc[euro12.Team.isin(['England', 'Italy', 'Russia']), ['Team','Shooting Accuracy']]
```
